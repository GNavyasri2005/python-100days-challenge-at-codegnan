```python
#using def function sum of 2 numbers
def sum_of_2_numbers(n1,n2):
    sum=n1+n2
    return sum
a=int(input("enter first number"))
b=int(input("enter second number"))
obj=sum_of_2_numbers(a,b)
print(obj)
```

```python
#using def function sum of 3 numbers
def sum_of_3_numbers(n1,n2,n3):
    sum=n1+n2+n3
    return sum
a=int(input("enter first number"))
b=int(input("enter second number"))
c=int(input("enter third number"))
obj=sum_of_3_numbers(a,b,c)
print(obj)
```

```python
#using def function sum of 3 numbers
def sum_of_4_numbers(n1,n2,n3,n4):
    sum=n1+n2+n3+n4
    return sum
a=int(input("enter first number"))
b=int(input("enter second number"))
c=int(input("enter third number"))
d=int(input("enter fourth number"))
obj=sum_of_4_numbers(a,b,c,d)
print(obj)
```

```python
#default arguments
def sum(a,b,c=0,d=0):
    return a+b+c+d
print(sum(1,2,3,4))
```

```python
#variable length of positional arguments
a,*b,c=10,20,30,40,50,21
print(f"a={a},b={b}")
```

```python
def vlpa(a,*b):
    return a,b
print(vlpa(1,2,3,4,5))
```
