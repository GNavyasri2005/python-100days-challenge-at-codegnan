```python
#simple calculator using functions
def addition(a,b):#parameters
    add=a+b
    return add
def subtraction(a,b):#parameters
    sub=a-b
    return sub
def multiplication(a,b):#parameters
    mul=a*b
    return mul
def division(a,b):#parameters
    div=a/b
    if b==0:
        return "ZERO DIVISION ERROR"
    return div  
obj1=addition(1,2)#arguments
obj2=subtraction(3,4)
obj3=multiplication(5,6)
obj4=division(7,8)
print(obj1)
print(obj2)
print(obj3)
print(obj4)
```

```python
def addition(a,b):#parameters
    add=a+b
    return add
def subtraction(a,b):#parameters
    sub=a-b
    return sub
def multiplication(a,b):#parameters
    mul=a*b
    return mul
def division(a,b):#parameters
    div=a/b
    if b==0:
        return "ZERO DIVISION ERROR"
    return div  
if __name__=="__main__":
    print("welcome to simple calculator")
    n1=int(input("enter a number1"))
    n2=int(input("enter a number2"))
    print("click 1 for addition \n click 2 for subtraction \n click 3 for multiplication \n click 4 for division")
    choice=int(input("enter your choice"))
    if choice==1:
        print(" addition of 2 numbers",addition(n1,n2))
    elif choice==2:
        print(" subtraction of 2 numbers",subtraction(n1,n2))  
    elif choice==3:
        print("multiplication of 2 numbers",multiplication(n1,n2))  
    elif choice==4:
        print(" division of 2 numbers",division(n1,n2))  
    else:
        print("enter valid choice")
    
```
