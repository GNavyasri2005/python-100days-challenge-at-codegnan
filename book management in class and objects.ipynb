```python
class BookManagementSystem:
    def __init__(self):
        self.books = {
            101: {"name": "Wings of Fire", "author": "A. P. J. Abdul Kalam", "quantity": 5},
            102: {"name": "The Alchemist", "author": "Paulo Coelho", "quantity": 4},
            103: {"name": "Rich Dad Poor Dad", "author": "Robert Kiyosaki", "quantity": 6},
            104: {"name": "Think and Grow Rich", "author": "Napoleon Hill", "quantity": 3},
            105: {"name": "Harry Potter and the Philosopherâ€™s Stone", "author": "J. K. Rowling", "quantity": 7},
            106: {"name": "The Power of Habit", "author": "Charles Duhigg", "quantity": 2},
            107: {"name": "Atomic Habits", "author": "James Clear", "quantity": 8},
            108: {"name": "Ikigai", "author": "Hector Garcia & Francesc Miralles", "quantity": 5},
            109: {"name": "The Monk Who Sold His Ferrari", "author": "Robin Sharma", "quantity": 4},
            110: {"name": "The Theory of Everything", "author": "Stephen Hawking", "quantity": 3}
        }
        self.book_id = set(self.books.keys())

    # Add book
    def add_book(self):
        id = int(input("Enter book id: "))
        if id in self.book_id:
            print("Book already exists")
        else:
            name = input("Enter book name: ")
            author = input("Enter author name: ")
            quantity = int(input("Enter quantity: "))

            self.book_id.add(id)
            self.books[id] = {
                "name": name,
                "author": author,
                "quantity": quantity
            }
            print("Book successfully added")

    # Update book
    def update_book(self):
        old_id = int(input("Enter book id to update: "))
        if old_id in self.book_id:
            new_id = int(input("Enter new id: "))
            if new_id in self.book_id and new_id != old_id:
                print("New ID already exists")
                return

            name = input("Enter new book name: ")
            author = input("Enter new author name: ")
            quantity = int(input("Enter new quantity: "))

            self.book_id.remove(old_id)
            del self.books[old_id]

            self.book_id.add(new_id)
            self.books[new_id] = {
                "name": name,
                "author": author,
                "quantity": quantity
            }
            print("Book updated successfully")
        else:
            print("Book not found")

    # Delete book
    def delete_book(self):
        id = int(input("Enter book id to delete: "))
        if id in self.book_id:
            self.book_id.remove(id)
            del self.books[id]
            print("Book successfully deleted")
        else:
            print("Book not found")

    # Total books
    def total_books(self):
        print("Total number of book titles:", len(self.book_id))

    # Display books
    def display_books(self):
        if not self.books:
            print("No books available")
        else:
            print("\n------ Book List ------")
            for id, details in self.books.items():
                print(f"ID: {id}")
                print(f"Name: {details['name']}")
                print(f"Author: {details['author']}")
                print(f"Quantity: {details['quantity']}")
                print("-" * 30)


def main():
    bms = BookManagementSystem()

    while True:
        print("\n1. Add Book")
        print("2. Update Book")
        print("3. Delete Book")
        print("4. Total Books")
        print("5. Display Books")
        print("6. Exit")

        choice = input("Enter your choice: ")

        if choice == "1":
            bms.add_book()
        elif choice == "2":
            bms.update_book()
        elif choice == "3":
            bms.delete_book()
        elif choice == "4":
            bms.total_books()
        elif choice == "5":
            bms.display_books()
        elif choice == "6":
            print("Thank you for using Book Management System ðŸ“š")
            break
        else:
            print("Invalid choice. Please try again.")


if __name__ == "__main__":
    main()
```
