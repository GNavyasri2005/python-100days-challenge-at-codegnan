```python
# N
for i in range(5):
    for j in range(5):
        if j == 0 or j == 4 or i == j:
            print("*", end=" ")
        else:
            print(" ", end=" ")
    print()
print()

# A
for i in range(5):
    for j in range(5):
        if j == 0 or j == 4 or i == 0 or i == 2:
            print("*", end=" ")
        else:
            print(" ", end=" ")
    print()
print()

# V
for i in range(5):
    for j in range(5):
        if (j == i and i < 4) or (j == 4 - i and i < 4):
            print("*", end=" ")
        else:
            print(" ", end=" ")
    print()
print()

# Y
for i in range(5):
    for j in range(5):
        if (i == j and i < 3) or (i + j == 4 and i < 3) or (j == 2 and i >= 2):
            print("*", end=" ")
        else:
            print(" ", end=" ")
    print()
print()

# A
for i in range(5):
    for j in range(5):
        if j == 0 or j == 4 or i == 0 or i == 2:
            print("*", end=" ")
        else:
            print(" ", end=" ")
    print()
```
