```python
             #############################SINGLE INHERITENCE #######################################
class Parent:  # SINGLE PARENT CLASS
    def __init__(self, mother: str, father: str, child: str):
        self.mother = mother
        self.father = father
        self.child = child
    def details(self):
        return f"Family: {self.mother}, {self.father}, {self.child}"
    def ages(self, mother_age: int, father_age: int, child_age: int):
        self.mother_age = mother_age
        self.father_age = father_age
        self.child_age = child_age
    def age_details(self):
        return f"Ages → Mother: {self.mother_age}, Father: {self.father_age}, Child: {self.child_age}"
class Child(Parent):  # SINGLE CHILD INHERITS PROPERTIES OF PARENT CLASS
    pass
s1 = Parent("Sita", "Ram", "LuvKush")
s2 = Child("Navya", "Shiv", "Mokshitha")
s1.ages(20, 21, 2)
s2.ages(25, 30, 5)

print(s1.details())
print(s1.age_details())
print(s2.details())
print(s2.age_details())
                               #parent---->child
```

```python
             ############################# MULTI-LEVEL INHERITENCE #######################################
class Parent:  # SINGLE PARENT CLASS
    def __init__(self, mother: str, father: str, child: str):
        self.mother = mother
        self.father = father
        self.child = child
    def details(self):
        return f"Family: {self.mother}, {self.father}, {self.child}"
    def ages(self, mother_age: int, father_age: int, child_age: int):
        self.mother_age = mother_age
        self.father_age = father_age
        self.child_age = child_age
    def age_details(self):
        return f"Ages → Mother: {self.mother_age}, Father: {self.father_age}, Child: {self.child_age}"
class Child(Parent):  # 1st CHILD INHERITS PROPERTIES OF PARENT CLASS
    pass
class grand_child(Child): # 2ND CHILD INHERITS PROPERTIES OF PARENT CLASS
    pass
s1 = Parent("Sita", "Ram", "LuvKush")
s2 = Child("Navya", "Shiv", "Mokshitha")
s3=grand_child("rohini","radhakrishna","luckymicky")
s1.ages(20, 21, 2)
s2.ages(25, 30, 5)
s3.ages(30,35,15)
print(s1.details())
print(s1.age_details())
print(s2.details())
print(s2.age_details())
print(s3.details())
print(s3.age_details())
                          #parent--->child----->grand_child
```
